```python
from confluent_kafka import avro
from confluent_kafka.avro import AvroProducer
from faker import Faker
import random
from datetime import datetime
from time import sleep
```

```python
lista_clientes = []
for r in range(1000):
    fake = Faker()
    lista_clientes.append({
        'id_cliente': r,
        'nome': fake.name(),
        'age': fake.random_int(min=1,max=99),
        'created_at': datetime.now().isoformat()
    })

lista_produtos = []
for r in range(1000):
    fake = Faker()
    lista_produtos.append({
        'id_produto': r,
        'nome_produto': fake.word(),
        'valor': fake.random_int(min=1.0, max=1000.0)
    })

lista_compra = []
for r in range(1000):
    lista_compra.append({
        'id_produto': random.sample(lista_produtos, 1)[0]['id_produto'],
        'id_cliente': random.sample(lista_clientes, 1)[0]['id_cliente']
    })
```

```python
kafka_conf = {
    'bootstrap.servers': 'kafka1:9092,kafka2:9092,kafka3:9092',
    'schema.registry.url': 'http://schema_registry:8100'
}

schema_cliente_str = """
{
   "namespace": "cadastro",
   "name": "cliente",
   "type": "record",
   "fields" : [
        {"name" : "id_cliente", "type" : "int"},
        {"name" : "nome", "type" : "string"},
        {"name" : "age", "type" : "int"},
        {"name" : "created_at", "type" : "string"}
   ]
}
"""
schema_cliente = avro.loads(schema_cliente_str)
avro_producer = AvroProducer(
    kafka_conf,
    default_value_schema=schema_cliente
)
for cliente in lista_clientes:
    avro_producer.produce(
        topic='topico_cliente',
        value=cliente)
    avro_producer.flush()


schema_produto_str = """
{
   "namespace": "mercado",
   "name": "produto",
   "type": "record",
   "fields" : [
        {"name" : "id_produto", "type" : "int"},
        {"name" : "nome_produto", "type" : "string"},
        {"name" : "valor", "type" : "int"}
   ]
}
"""
schema_produto = avro.loads(schema_produto_str)
avro_producer = AvroProducer(
    kafka_conf,
    default_value_schema=schema_produto
)
for produto in lista_produtos:
    avro_producer.produce(
        topic='topico_produto',
        value=produto)
    avro_producer.flush()


schema_compra_str = """
{
   "namespace": "mercado",
   "name": "compra",
   "type": "record",
   "fields" : [
        {"name" : "id_produto", "type" : "int"},
        {"name" : "id_cliente", "type" : "int"}
   ]
}
"""
schema_compra = avro.loads(schema_compra_str)
avro_producer = AvroProducer(
    kafka_conf,
    default_value_schema=schema_compra
)
for compra in lista_compra:
    avro_producer.produce(
        topic='topico_compra',
        value=compra)
    avro_producer.flush()
```
