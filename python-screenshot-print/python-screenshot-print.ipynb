```python
from mss import mss
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

def display_image(grab_image, size=(10,15)):
    plt.figure(figsize=size)
    PilImage = Image.frombytes("RGB", grab_image.size, grab_image.bgra, "raw", "BGRX")
    plt.imshow(PilImage)

cap = mss()

n_monitor = 1
monitor = cap.monitors[n_monitor]
location = {'top': monitor['top'] + 0,
            'left': monitor['left'] + 0,
            'width': monitor['width'] - 2300,
            'height': monitor['height'] - 1000,
            'mon': 1}

grabed_image = cap.grab(location)

display_image(grabed_image)
```
